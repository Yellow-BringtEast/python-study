```python
# 增加列
# 增加一个固定值的列
df['one'] = 1
df
```

```python
# 增加总成绩列
df['total'] = df.Q1 + df.Q2 + df.Q3 + df.Q4
df
```

```python
# 将计算得来的结果赋值给新列
df['total'] = df.loc[:, 'Q1':'Q4'].apply(lambda x:sum(x), axis=1)
df
```

```python
df['avg'] = df.total / 4
df
```
