```python
import pandas as pd

df = pd.read_excel('./team.xlsx')
```

```python
# 函数筛选
df.Q1[lambda s: max(s.index)]
```

```python
# 查询
df.query('Q1 > Q2 > 90')
```

```python
df.query('team != "C"')
```

```python
a = df.Q1.mean()
df.query('Q1 > @a+40')
```

```python
# 筛选
df.filter(items=['Q1', 'Q2'])
```

```python
df.filter(regex='Q', axis=1)
```

```python
# 智能推断类型
df.infer_objects().dtypes
```

```python
df.convert_dtypes().dtypes
```

```python
# 高级过滤
df = df.select_dtypes(include='number')
df.where(df > 70)
```

```python
df.where(lambda d: d.Q1>50)
```

```python
df.where(df>=60, "不及格")
```

```python
c = df%2 == 0
df.where(~c, -(df-20))
```

```python
# 数据迭代
for name in df.name:
    print(name)
```

```python
for index, row in df.iterrows():
    print(index, row['name'], row.Q1)
```

```python
for column in df:
    print(column)
```

```python
# pipe()
df.pipe(lambda df_, x, y: df_[(df_.Q1 >= x) & (df.Q2 >= y)], 80, 90)
```

```python
# apply()
df.name.apply(lambda x: x.lower())
```

```python
def my_mean(s):
    max_min_ser = pd.Series([-s.max(), -s.min()])
    return s.append(max_min_ser).sum() / (s.count() - 2)

df.select_dtypes(include='number').apply(my_mean)
```

```python
# agg()
df.agg('max')
```
