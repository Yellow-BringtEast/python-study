```python
import pandas as pd

df = pd.read_excel('./team.xlsx')
```

```python
# 函数筛选
df.Q1[lambda s: max(s.index)]
```

```python
# 查询
df.query('Q1 > Q2 > 90')
```

```python
df.query('team != "C"')
```

```python
a = df.Q1.mean()
df.query('Q1 > @a+40')
```

```python
# 筛选
df.filter(items=['Q1', 'Q2'])
```

```python
df.filter(regex='Q', axis=1)
```

```python
# 智能推断类型
df.infer_objects().dtypes
```

```python
df.convert_dtypes().dtypes
```

```python
# 高级过滤
df = df.select_dtypes(include='number')
df.where(df > 70)
```

```python
df.where(lambda d: d.Q1>50)
```

```python
df.where(df>=60, "不及格")
```

```python
c = df%2 == 0
df.where(~c, -(df-20))
```
