```python
import pandas as pd

df = pd.read_excel('./team.xlsx')
```

```python
# 数据追加
df1 = pd.DataFrame({'x': [1, 2], 'y': [3, 4]})
df1
```

```python
df2 = pd.DataFrame({'x': [5, 6], 'y': [7, 8]})
df2
```

```python
# 追加合并
df1.append(df2)
```

```python
# 追加多个数据
df1.append([df2, df2])
```

```python
df3 = pd.DataFrame({'y': [5, 6], 'z': [7, 8]})
df3
```

```python
df1.append(df3)
```

```python
# 追加序列
# 定义新同学的信息
lily = pd.Series(['lily', 'C', 55, 56, 57, 58], index=[
                 'name', 'team', 'Q1', 'Q2', 'Q3', 'Q4'])
lily
```

```python
df = df.append(lily, ignore_index=True)

df.tail()
```

```python
# 追加字典
# 将学生信息定义为一个字典
lily = {'name': 'lily', 'team': 'C', 'Q1': 55, 'Q2': 56, 'Q3': 57, 'Q4': 58}
df = df.append(lily, ignore_index=True)

df.tail()
```

```python
# 数据连接
df2 = pd.DataFrame({'x': [5, 6, 0], 'y': [7, 8, 0]})
df2

# 按列连接
pd.concat([df1, df2], axis=1)
```

```python
# 按列合并交集
pd.concat([df1, df2], axis=1, join='inner')
```

```python
# 指定索引名
pd.concat([df1, df2], keys=['a', 'b'])
```

```python
# 横向合并，指定索引
pd.concat([df1, df2], axis=1, keys=['a', 'b'])
```

```python
# 数据合并
df1 = pd.DataFrame({'a': [1, 2], 'x': [5, 6]})
df1
```

```python
df2 = pd.DataFrame({'a': [2, 1, 0], 'y': [6, 7, 8]})
df2
```

```python
pd.merge(df1, df2, on='a')
```

```python
# 索引连接
pd.merge(df1, df2,
         left_index=True,
         right_index=True,
         suffixes=('_1', '_2')
         )
```

```python
# 多连接键
df3 = pd.DataFrame({'a': [1, 2], 'b': [3, 4], 'x': [5, 6]}) 
df3
```

```python
df4 = pd.DataFrame({'a': [1, 2, 3], 'b': [3, 4, 5], 'y': [6, 7, 8]})
df4
```

```python
pd.merge(df3, df4, on=['a','b'])
```

```python
# 以左表为基表
pd.merge(df3, df4, how='left', on=['a', 'b'])
```

```python
# 以右表为基表
pd.merge(df3, df4, how='right', on=['a', 'b'])
```

```python
# 一个有重复连接键的例子
left = pd.DataFrame({'A': [1, 2], 'B': [2, 2]}) 
right = pd.DataFrame({'A': [4, 5, 6], 'B': [2, 2, 2]}) 
pd.merge(left, right, on='B', how='outer')
```

```python
# 显示连接指示列
pd.merge(df1, df2, on='a', how='outer', indicator=True)
```
