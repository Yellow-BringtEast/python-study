```python
import pandas as pd

df = pd.read_excel('./team.xlsx')
```

```python
# 分组
df.groupby('team').sum()
```

```python
# 分组后不同的聚合
df.groupby('team').agg({'Q1': 'sum', 'Q2': 'count', 'Q3': 'mean', 'Q4': 'max'})
```

```python
# 分组对象
df.groupby('team')
```

```python
# 通过表达式分组
df.groupby(lambda x: x % 2 == 0).sum()
```

```python
# 使用pipe调用分组函数
df.pipe(pd.DataFrame.groupby, 'team').sum()
```

```python
# 分组器Grouper
df.groupby(pd.Grouper('team')).sum()
```
